```python
# Параметри задачі

sample_mean = np.mean(data)
population_mean = 20
population_std = np.std(data, ddof=1)
sample_size = len(data)
alpha = 0.05
```

Гіпотези

H₀: μ = 20 (середній чек дорівнює $20)

H₁: μ > 20 (середній чек вище ніж $20)

Тип тесту: правосторонній (one-tailed)

```python
# Обчислюємо стандартну помилку

standard_error = population_std / np.sqrt(sample_size)
print(f"SE = {standard_error:.3f}")
```

```python
# Розраховуємо z-статистику

z_statistic = (sample_mean - population_mean) / standard_error
print(f"z = {z_statistic:.3f}")
```

```python
# Знаходимо критичне z-значення

z_critical = stats.norm.ppf(1 - alpha)
print(f"Критичне z-значення = {z_critical:.3f}")
```

```python
# Обчислюємо p-value

p_value = 1 - stats.norm.cdf(z_statistic)
print(f"p-value = {p_value:.3f}")
```

Чи зійшлись значення z-статистику та р-значення в цьому завданні з попередніми обчисленнями?

```python
from statsmodels.stats import weightstats as stests
```

```python
# Виконуємо правосторонній z-тест

z_stat_sm, p_val_sm = stests.ztest(data,
                                   value=population_mean,
                                   alternative='larger')

print(f"Результати z-тесту:")
print(f"Z-статистика: {z_stat_sm:.3f}")
print(f"P-value: {p_val_sm:.3f}")
```

Значення z-статистики та p-value зійшлись з результатами попередніх обчислень.
